```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr = np.arange(10)
```

```python
arr
```

```python
arr_sq = arr**2
```

```python
arr_sq
```

# Line Graph

```python
plt.plot(arr ,arr, marker="^", c="red", label="y=x")
plt.show()

plt.plot(arr, arr_sq, marker="*", c="black", label="y= x**2")
plt.legend()
plt.xlabel("Numbers (0-10)")
plt.ylabel("Range of function")
plt.title("Comparision of Functions")
plt.xlim(1, 20 )
# plt.ylim()
plt.xticks(np.arange(0, 20, step=1))
plt.show()
```

# Scatter Plot

```python
data1 = np.random.randn(100, 2)
data2 = 5*np.random.randn(100, 2) + 10
data1_x = data1[:,0]
data1_y = data1[:,1]

data2_x = data2[:,0]
data2_y = data2[:,1]

plt.scatter(data1_x, data1_y, label= "small range", marker="^")
plt.scatter(data2_x, data2_y, label= "big range")
plt.legend()
plt.show()
```

```python
data1[:,0].mean()
```

```python
data2[:, 0].mean()
```

## Bar plot


```python
x = np.array([1,2,3])
sales_2019 = [20,5,13]
sales_2020 = [12, 10, 21]
```

```python
plt.bar(x, sales_2019, width=0.2, align='edge')
plt.bar(x+0.2, sales_2020, width=0.2, align='edge')
plt.show()
```
